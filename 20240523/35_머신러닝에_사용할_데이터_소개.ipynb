```python
import warnings
warnings.filterwarnings(action='ignore')
%config Completer.use_jedi = False
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.rcParams['font.size'] = 10
import seaborn as sns
```

```python
# 사이킷런 라이브러리에서 제공하는 데이터 셋을 사용하기 위해 import 한다.
from sklearn import datasets
```

사이킷런 라이브러리를 사용한 머신러닝에서 사용할 데이터 소개

보스턴 집값 예측하기

```python
# 사이킷런 라이브러리에서 제공하는 보스턴 집값 데이터를 불러온다.
# 사이킷런에서 제공하는 데이터는 data(피쳐) 부분과 target(레이블, 클래스, 실제값) 부분으로 구성되어 있다.
# feature_names(피쳐 이름), DESCR(피쳐의 설명), filename(csv 파일 이름), data_module(데이터 모듈이름) 정보를 제공한다.
raw_data = datasets.load_boston()
raw_data
```

```python
# 피쳐 부분을 추출해서 데이터프레임을 만든다.
x_train = pd.DataFrame(raw_data.data)
print(x_train.shape)
x_train.head()
```

```python
# 레이블 부분을 추출해서 데이터프레임을 만든다.
y_train = pd.DataFrame(raw_data.target)
print(y_train.shape)
y_train.head()
```

```python
# 피쳐와 레이블 데이터를 결합해서 전체 데이터프레임을 생성한다.
# 판다스의 concat() 메소드를 이용해서 데이터프레임을 합칠 수 있다.
# concat() 메소드의 axis 속성으로 데이터프레임의 결합 방향을 지정할 수 있다.
# axis 속성을 생략하면 0이 기본값이고 데이터프레임을 위에서 아래 방향으로 합치고 1로 설정하면
# 왼쪽에서 오른쪽으로 합친다.
raw_data_df = pd.concat([x_train, y_train], axis=1)
print(raw_data_df.shape)
raw_data_df.head()
```

```python
# 피쳐와 레이블을 결합한 데이터프레임의 열 이름을 지정한다.
feature_names = raw_data.feature_names
print(type(feature_names))
print(feature_names)
columns = np.append(feature_names, ['label'])
print(columns)
raw_data_df.columns = columns
raw_data_df.head()
```

```python
# 필요하다면 작업이 완료된 데이터프레임을 csv 파일로 저장한다.
raw_data_df.to_csv('./data/sklearn/boston.csv', index=False)
```
