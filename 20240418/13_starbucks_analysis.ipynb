```python
import requests
import folium
# 파이썬 데이터 처리 라이브러리 pandas를 설치하고 import 한다.
# !pip install pandas
import pandas as pd
# 파이썬의 딕셔너리 타입의 데이터를 pandas 데이터 프레임으로 변환하기위해 import 한다.
from pandas.io.json import json_normalize
```

전국 또는 특정 지역의 스타벅스 매장 위치 정보를 가져와서 지도위에 시각화하기

requests 모듈로 스타벅스 매장 위치 정보를 가져와서 딕셔너리 타입으로 변경한다.  
json_normalize() 함수를 사용해서 json 타입의 데이터가 변환된 딕셔너리를 데이터프레임으로 변환한다.  
folium 모듈을 사용해서 지도를 만들고 지도위에 스타벅스 매장 위치를 marker로 표시한다.

```python
# 서울특별시 종로구 스타벅스 전 지점위치 크롤링
targetSite = 'https://www.starbucks.co.kr:7643/store/getStore.do?r=6AIKXR4SA9'
request = requests.post(targetSite, data={
    'ins_lat': 37.563398,
    'ins_lng': 126.9863309,
    'p_sido_cd': '',
    'p_gugun_cd': '0118',
    'in_biz_cd': '', 
    'iend': 2000, 
    'set_date': ''
})
starbucks = request.json()
store = starbucks['list']
for i in range(len(store)):
    print('{:4d}. {}: {}({}, {})'.format(i + 1, store[i]['s_name'], store[i]['doro_address'], store[i]['lat'], store[i]['lot']))
```

```python
# pandas.io.json 모듈의 json_normalize() 함수로 json 타입의 데이터가 변환된 딕셔너리를 판다스
# 데이터프레임으로 변환한다.
# json_normalize(딕셔너리, '데이터프레임으로 변환할 데이터가 할당된 딕셔너리의 key')
starbucks_df = json_normalize(starbucks, 'list')
print(type(starbucks_df)) # <class 'pandas.core.frame.DataFrame'>
print(len(starbucks_df))
print(starbucks_df.columns)
starbucks_df
```

딕셔너리를 데이터프레임으로 변환하니 서울특별시 종로구에 있는 스타벅스 지점은 40개고 각 지점당 134개의 컬럼으로 정보를 표시하고 있다.  
작업에 필요한 컬럼 몇 가지를 선택해서 지도에 마커를 표시할 때 사용할 데이터프레임을 만든다.  

s_name => 지점이름  
sido_code => 시도코드  
sido_name => 시도이름  
gugun_code => 구군코드  
gugun_name => 구군이름  
doro_address => 도로명주소  
lat => 위도  
lot => 경도

```python
# 판다스 데이터프레임에서 특정 컬럼의 데이터를 얻어올 때 얻어올 컬럼이 1개라면 컬럼 이름만 적어도 되지만
# 얻어올 컬럼이 2개 이상일 경우 얻어올 컬럼들을 반드시 리스트 형태로 만들어서 얻어와야 한다.
starbucks_df_map = starbucks_df[
    ['s_name', 'sido_code', 'sido_name', 'gugun_code', 'gugun_name', 'doro_address', 'lat', 'lot']
]
del starbucks_df
starbucks_df_map
```

```python
# dtypes 속성으로 데이터프레임을 구성하는 컬럼들의 자료형을 확인할 수 있다.
# 판다스는 문자열 데이터 타입을 object라 부른다.
starbucks_df_map.dtypes
```

```python
# info() 메소드를 사용해도 데이터프레임을 구성하는 컬럼들의 자료형을 확인할 수 있다.
starbucks_df_map.info()
```

```python
# astype() 메소드로 위도, 경도 데이터 타입을 object에서 float로 변환한다.
starbucks_df_map['lat'] = starbucks_df_map['lat'].astype(float)
starbucks_df_map['lot'] = starbucks_df_map['lot'].astype(float)
starbucks_df_map.info()
```

```python
foliumMap = folium.Map(location=[37.56976899, 126.98365027], zoom_start=17)
popup = folium.Popup('종각점', max_width=300)
folium.Marker(location=[37.569918, 126.984528], popup=popup).add_to(foliumMap)
popup = folium.Popup('종로관철점', max_width=300)
folium.Marker(location=[37.569058, 126.986013], popup=popup).add_to(foliumMap)
popup = folium.Popup('종로구청점', max_width=300)
folium.Marker(location=[37.572662, 126.981138], popup=popup).add_to(foliumMap)
foliumMap.save('./output/starbucksMap1.html')
foliumMap
```

```python
# 서울특별시 종로구에 위치한 스타벅스 전 지점의 위치를 지도위에 표시한다.
# 지도를 작성할 때 지도의 중심을 설정할 위치가 애매할 경우 지도상에 표시할 모든 스타벅스 지점의
# 위도와 경도의 평균을 중심으로 하는 지도를 만들어 사용하면 된다.
# 데이터프레임에 저장된 데이터의 평균은 mean() 메소드를 사용해서 계산한다.
foliumMap = folium.Map(location=[starbucks_df_map['lat'].mean(), starbucks_df_map['lot'].mean()], zoom_start=14)

# iterrows() 메소드는 데이터 프레임에 저장된 데이터의 인덱스와 데이터를 튜플 형태로 리턴한다.
for index, data in starbucks_df_map.iterrows():
    # print(index, data)
    popup = folium.Popup('{}. {} - {}'.format(index + 1, data['s_name'], data['doro_address']), max_width=300)
    folium.Marker(location=[data['lat'], data['lot']], popup=popup).add_to(foliumMap)

foliumMap.save('./output/starbucksMap2.html')
foliumMap
```

```python
# 서울특별시 강남구에 위치한 스타벅스 전 지점의 위치를 지도위에 표시한다.
targetSite = 'https://www.starbucks.co.kr:7643/store/getStore.do?r=6AIKXR4SA9'
request = requests.post(targetSite, data={
    'ins_lat': 37.563398,
    'ins_lng': 126.9863309,
    'p_sido_cd': '',
    'p_gugun_cd': '0101',
    'in_biz_cd': '', 
    'iend': 2000, 
    'set_date': ''
})
starbucks = request.json()
starbucks_df = json_normalize(starbucks, 'list')
starbucks_df_map = starbucks_df[['s_name', 'sido_code', 'sido_name', 'gugun_code', 'gugun_name', 'doro_address', 'lat', 'lot']]
starbucks_df_map['lat'] = starbucks_df_map['lat'].astype(float)
starbucks_df_map['lot'] = starbucks_df_map['lot'].astype(float)

foliumMap = folium.Map(location=[starbucks_df_map['lat'].mean(), starbucks_df_map['lot'].mean()], zoom_start=14)
for index, data in starbucks_df_map.iterrows():
    popup = folium.Popup('{}. {} - {}'.format(index + 1, data['s_name'], data['doro_address']), max_width=300)
    folium.Marker(location=[data['lat'], data['lot']], popup=popup).add_to(foliumMap)
foliumMap.save('./output/starbucksMap3.html')
foliumMap
```
