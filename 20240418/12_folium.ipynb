```python
# folium 라이브러리를 사용하면 지도를 이용한 시각화를 할 수 있다.
# folium 라이브러리를 설치하고 import 한다.
# !pip install folium
import folium
```

```python
# folium 라이브러리의 Map 클래스로 location 속성과 zoom_start 속성을 지정해서 지도를 만든다.
# location 속성으로 위도와 경도를 지정해서 지도의 중심을 설정한다.
# zoom_start 속성으로 지도의 배율을 설정한다.
foliumMap = folium.Map(location=[37.56976899, 126.98365027], zoom_start=18)

# folium 라이브러리의 Market 클래스로 location 속성으로 지정한 위치에 popup 속성으로 지정한 문자열을
# 표시하는 marker를 만든다.
# Market 클래스로 작성한 marker는 add_to() 메소드로 지도에 추가한다.
# folium.Marker(location=[37.5699183, 126.98452810000003], popup='스타벅스 종로점').add_to(foliumMap)

# marker를 클릭했을 때 팝업되는 말풍선 크기가 마음에 들지 않으면 folium 라이브러리의 Popup 클래스로
# 말풍선의 최대 가로 크기를 지정할 수 있다.
popup = folium.Popup('스타벅스 종로점', max_width=300)
folium.Marker(location=[37.5699183, 126.98452810000003], popup=popup).add_to(foliumMap)
popup = folium.Popup('스타벅스 종로관철점', max_width=300)
folium.Marker(location=[37.569058, 126.986013], popup=popup).add_to(foliumMap)

# save() 메소드로 folium을 사용해서 작성한 지도를 html 파일로 저장할 수 있다.
foliumMap.save('./output/foliumMap.html')
foliumMap
```
