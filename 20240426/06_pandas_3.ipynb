분석하기 좋은 데이터

분석하기 좋은 데이터는 데이터 집합을 분석하기 좋은 형태로 만들어 놓은 것을 말하고 데이터 분석 작업에서 데이터 정리 아주 중요한 작업으로 실제 데이터 분석 작업의 70% 이상을 차지하고 있는 작업이 데이터 정리 작업이다.

concat() 메소드는 데이터프레임을 연결한다.  
연결하려는 데이터프레임을 리스트 담아서 cancat() 메소드에 전달하면 연결한 데이터프레임을 리턴한다.  
concat([데이터프레임1, 데이터프레임2, ...][, axis])  
axis 속성을 생략하면 0이 기본값이고 위에서 아래 방향으로 연결하고 axis=1로 지정하면 왼쪽에서 오른쪽 방향으로 연결한다.

```python
# df1, df2, df3 데이터프레임은 열 인덱스가 모두 A, B, C, D로 같아서 연결한 다음에도 열이 그대로 유지된다.
row_concat = pd.concat([df1, df2, df3]) # pd.concat([df1, df2, df3], axis=0), 위에서 아래 방향으로 연결
row_concat
```

```python
# df1, df2, df3 데이터프레임은 행 인덱스가 모두 0, 1, 2, 3으로 같아서 연결한 다음에도 행이 그대로 유지된다.
col_concat = pd.concat([df1, df2, df3], axis=1) # 왼쪽에서 오른쪽 방향으로 연결
col_concat
```

```python
# 같은 열 인덱스가 존재하는 데이터프레임에서 열 인덱스로 추출하면 해당 열 인덱스의 데이터가 모두 추출된다.
col_concat.A
```

append() 메소드를 사용하면 데이터프레임을 사용하지 않고 딕셔너리를 사용해서 데이터프레임에 추가할 수 있다.

```python
data_dict = {'A': 'n5', 'B': 'n6', 'C': 'n7', 'D': 'n8'}
print(data_dict)
# concat() 메소드의 인수는 시리즈나 데이터프레임만 지정할 수 있기 때문에 에러가 발생된다.
# pd.concat([df1, data_dict]) # 에러
# 딕셔너리는 인덱스가 없으므로 ignore_index 속성을 이용해서 인덱스를 무시하게 설정하면 
# 데이터프레임에 딕셔너리를 추가할 수 있다.
# df1.append(data_dict) # 에러
df1.append(data_dict, ignore_index=True)
```

사용중인 데이터프레임의 열 인덱스를 변경하려면 columns 속성에 열 이름을 리스트 형태로 전달한다.

```python
print(df1.columns) # columns 속성은 열 인덱스 목록을 기억하고 있다.
df2.columns = ['E', 'F', 'G', 'H']
print(df2.columns)
df3.columns = ['A', 'C', 'F', 'H']
print(df3.columns)
```

```python
# 새롭게 데이터프레임의 열 인덱스를 부여하고 concat() 메소드로 연결하면 데이터프레임에 없는 열 인덱스의
# 데이터는 NaN(누락값)으로 처리된다.
pd.concat([df1, df2, df3], ignore_index=True)
```

concat() 메소드의 join 속성을 사용하면 연결하는 데이터프레임에 공통적으로 포함된 열만 골라서 연결할 수 있다.

```python
# join 속성의 기본값은 outer이고 공통으로 데이터가 있는 열에 신경쓰지 않고 연결한다. => NaN이 대량 발생된다.
pd.concat([df1, df2, df3], ignore_index=True, join='outer')
```

```python
# df1, df2, df3 데이터프레임은 3개 모두 공통 열 인덱스가 없기 때문에 빈 데이터프레임이 생성된다.
pd.concat([df1, df2, df3], ignore_index=True, join='inner')
```

```python
# df1, df3 데이터프레임은 2개 모두 공통 열 인덱스가 있기 때문에 공통 열을 연결한 데이터프레임이 생성된다.
pd.concat([df1, df3], ignore_index=True, join='inner')
```

```python
# df2, df3 데이터프레임은 2개 모두 공통 열 인덱스가 있기 때문에 공통 열을 연결한 데이터프레임이 생성된다.
pd.concat([df2, df3], ignore_index=True, join='inner')
```

merge() 메소드는 2개의 데이터프레임을 병합한다.  
기본값으로 내부 조인을 실행하며 메소드를 실행한 데이터프레임이 left로 지정되고 메소드의 인수로 지정한 데이터프레임이 right로 지정된다.  
left_on 속성과 right_on 속성은 각각 왼쪽과 오른쪽 데이터프레임에 일치해야 할 열 인덱스를 지정하고 how 속성으로 두 데이터프레임을 병합하는 방법을 지정한다.  
왼쪽데이터프레임.merge(오른쪽데이터프레임, left_on, right_on[, how])

```python
# survey_person.csv => 조사한 사람
# survey_site.csv => 조사 위치
# survey_survey.csv => 기상 정보
# survey_visited.csv => 조사한 날짜
```

```python
person = pd.read_csv('./data/survey_person.csv')
person
```

```python
site = pd.read_csv('./data/survey_site.csv')
site
```

```python
survey = pd.read_csv('./data/survey_survey.csv')
survey
```

```python
visited = pd.read_csv('./data/survey_visited.csv')
visited
```

```python
# site 데이터프레임과 visited 데이터프레임 merge
print(site.columns)
print(visited.columns)

# how 속성을 두 데이터프레임을 병합하는 방법을 지정한다. 기본값은 how='inner' 이다.
# how 속성을 생략하거나 how='inner'(내부 조인)로 지정하면 두 데이터프레임에 공통적으로 포함된 값이 있는 데이터만 병합된다.
# how 속성을 how='outer'(외부 조인)로 지정하면 두 데이터프레임의 모든 데이터가 병합된다. 한쪽에만 있는 데이터에는 NaN이 채워진다.
# how 속성을 how='left'(왼쪽 외부 조인)로 지정하면 왼쪽 데이터프레임의 모든 데이터와 오른쪽 데이터프레임에 있는 공통적으로
# 포함된 데이터가 병합된다. 한쪽에만 있는 데이터에는 NaN이 채워진다.
# how 속성을 how='outer'(오른쪽 외부 조인)로 지정하면 오른쪽 데이터프레임의 모든 데이터와 왼쪽 데이터프레임에 있는 공통적으로
# 포함된 데이터가 병합된다. 한쪽에만 있는 데이터에는 NaN이 채워진다.

# site.merge(visited, left_on='name', right_on='site')
# site.merge(visited, left_on='name', right_on='site', how='inner')
# site.merge(visited, left_on='name', right_on='site', how='outer')
# site.merge(visited, left_on='name', right_on='site', how='left')
site.merge(visited, left_on='name', right_on='site', how='right')
```

```python
# person 데이터프레임과 survey 데이터프레임 merge
print(person.columns)
print(survey.columns)
p2s = person.merge(survey, left_on='ident', right_on='person', how='inner')
p2s
```

```python
# visited 데이터프레임과 survey 데이터프레임 merge
print(visited.columns)
print(survey.columns)
v2s = visited.merge(survey, left_on='ident', right_on='taken', how='inner')
v2s
```

```python
# p2s 데이터프레임과 v2s 데이터프레임 merge
print(p2s.columns)
print(v2s.columns)
# left_on 속성과 right_on 속성에 리스트로 묶어서 여러개의 열 인덱스를 전달할 수 있다.
p2s.merge(v2s, left_on=['ident', 'taken', 'quant', 'reading'], right_on=['person', 'taken', 'quant', 'reading'], how='inner')
# merge() 메소드를 실행하는 두 데이터프레임에 공통된 열 인덱스가 있으면 왼쪽 데이터프레임의 열 인덱스에는 '_x'라는
# 접미사가 자동으로 추가되고 오른쪽 데이터프레임의 열 인덱스에는 '_y'라는 접미사가 자동으로 추가된다.
```
